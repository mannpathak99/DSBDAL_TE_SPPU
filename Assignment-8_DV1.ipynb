```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
plt.figure(figsize=(3,4))
sns.histplot(data=df,x='fare')
```

```python
sns.histplot(data=df,x='fare',binwidth=30)
```

```python
sns.histplot(data=df,x='fare',binwidth=30,hue='pclass')
```

```python
sns.displot(df["fare"],aspect=2.5)
```

```python
sns.displot(df["fare"],aspect=2.5,binwidth=20)
```

```python
sns.displot(data=df,x="age",hue="sex")
```

```python
sns.displot(data=df,x="age",hue="sex",col="pclass")
```

```python
sns.displot(data=df,x="age",hue="alive")
```

```python
sns.displot(data=df,x="sex",hue="alive")
```

```python
sns.barplot(x='sex',y='age',data=df)
```

```python
sns.barplot(x='sex',y='age',data=df,hue="alive")
```

```python
sns.barplot(x='pclass',y="age",data=df,hue="embarked",ci=None)
```

```python
sns.countplot(x='pclass',data=df)
```

```python
sns.countplot(x='pclass',data=df,hue="sex")
```

```python
sns.countplot(x='pclass',data=df,hue="sex", dodge=False)
```

```python
sns.catplot(x="embarked",data=df,hue='survived',kind="count")
```
